```python
# Basic Data Manipulation and Analysis
import pandas as pd  # Pandas for data manipulation
import numpy as np   # NumPy for numerical operations

# Data Visualization
import seaborn as sns  # Seaborn for statistical data visualization
import matplotlib.pyplot as plt  # Matplotlib for general plotting

# Clustering Methods
from sklearn.cluster import KMeans  # K-Means clustering
from sklearn.preprocessing import StandardScaler  # Standardization for feature scaling
from sklearn.decomposition import PCA  # Principal Component Analysis for dimensionality reduction
from sklearn.metrics import silhouette_score  # Silhouette score for evaluating cluster quality
```

```python
df = pd.read_csv("candy-data.csv")
```

```python
subset = df.iloc[:, 1:-3] # No first or last 3 columns.
subset
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(subset)
```

```python
pca.explained_variance_ratio_
# represents around 60% of the data.
```

```python
candy_2d = pd.DataFrame(pca.transform(subset))
candy_2d.head()
candy_2d.shape
```

```python
candy_2d.plot(kind="scatter",x=0,y=1)
```

```python
df["x_jitter"] = candy_2d[0] + np.random.randn(85)*0.1
df["y_jitter"] = candy_2d[1] + np.random.randn(85)*0.1

df.plot(kind="scatter",x="x_jitter",y="y_jitter")
```

```python
df.to_csv("canda_data_pca.csv")
```
